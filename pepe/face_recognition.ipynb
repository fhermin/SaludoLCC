```python


# Abre la cÃ¡mara de video
cap = cv2.VideoCapture(0)  # 0 para la cÃ¡mara predeterminada

# Inicializa un contador de frames
frame_count = 0

# Main loop
while True:
    # Captura un frame del video
    state, frame = cap.read()

    if not state:
        break

    frame_count += 1

    if frame_count == 6:
        res = DeepFace.find(frame, db_path='./Database/', enforce_detection=False, model_name='Facenet512')
        
        if len(res[0]['identity']) > 0:
            name = res[0]['identity'][0].split('/')[1].split('\\')[1]
            print(name)
            # Dibuja el nombre en el frame
            cv2.putText(frame, name, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)


        # Reinicia el contador de frames
        frame_count = 0

    # Muestra el frame
    cv2.imshow('Attendance', frame)

    key = cv2.waitKey(1) & 0xFF

    # Rompe el ciclo si se presiona la tecla 'q'
    if key == ord('q'):
        break

# Libera la cÃ¡mara y cierra todas las ventanas de OpenCV
cap.release()
cv2.destroyAllWindows()

```
